### Big O Notation

Big o notation is used to measure how running time or space req for ur program grows as input grows
<br> take this next algo growing as an example :

In this example the time that it takes to excute the programme increases equally to the increase of number of inputs
<br> A code example of this :

```python
def getSquared(numbers) :
    squaredNumbers = []
    for num in numbers :
        squaredNumbers.append(num * num)
    return squaredNumbers
```

If we increase the number of inputs in this case by 10 the time of excution will increase also by 10
<br> So we call it LINEAR GROWING and time = O(n)
