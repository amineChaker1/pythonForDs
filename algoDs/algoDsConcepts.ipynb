### Big O Notation

Big o notation is used to measure how running time or space req for ur program grows as input grows
<br> take this next algo growing as an example :

TIME COMPLEXITY OF O(n)

In this example the time that it takes to excute the programme increases equally to the increase of number of inputs
<br> A code example of this :

```python
def getSquared(numbers) :
    squaredNumbers = []
    for num in numbers :
        squaredNumbers.append(num * num)
    return squaredNumbers
```

If we increase the number of inputs in this case by 10 the time of excution will increase also by 10
<br> So we call it LINEAR GROWING and time = O(n)

TIME COMPLEXITY OF O(1)

in this example the excution time doesnt change even if the input size changes
<br> example for this :

```python
def getPip(prices, stocks,index):
    # index in this case is an integer
    pip = prices[index] / stocks[index]
    return pip
```

with this code it does the same math operation and it doesnt matter if the number of prices and stocks taken changes it will stay with the same time complexity O(1)

THE COMPLEXITY OF O(n2)

When we repeat the time complexity of O(n) will we will get the time com of O(n2)

So if your programme has diffrent time complexity mesurments we only keep the bigger one :
