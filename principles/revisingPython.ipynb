```python
# this is to append the first small number
R = [1,2,4,-5,7,9,3,2]
IR = []
j = 0
min = R[j]
for i in R :
    if R[j] < i :
        j += 1
        if R[j] < min :
         min = R[j]
         IR.append(min)
    else:
        continue
print(IR)
```

```python
# this is to change the order of the first smaller number
R = [1,2,4,-5,7,9,3,2]
min = R[0]
c = 0 
idx = 0

for i in R:
    if i < min:
        min = i
        idx = c
    c += 1
temp = R[0]
R[0] = R[idx]
R[idx] = temp
print(R)
```

```python
# this is to change the order of the whole list using loops
R = [1,2,4,-5,7,9,3,2]
for j in range(len(R)):
    m = R[j]
    c = j
    idx = j
    for i in range(j , len(R)):
        if R[i] < m :
            m = R[i]
            idx = c
        c+=1
    temp = R[j]
    R[j] = m
    R[idx] = temp
print(R)      
```

```python
# now how to handle an arbitrary number of arguments
# use the star * with the name of the argument and it will make the argument as a list
def addFunction(*args):
    sum = 0
    for i in range(len(args)):
     sum+= args[i]
    return sum
addFunction(2,5,6,7)
```
